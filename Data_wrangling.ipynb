**Let us create some subsets of our dataset**

```python
fuel_1 = df[df['fuel']=='Petrol']
fuel_2 = df[df['fuel']=='Petrol/CNG']
fuel_3 = df[df['fuel']=='nan']
```

**Concat both axis, default axis = 0(vertical axis)**

```python
all_fuel = pd.concat([fuel_1, fuel_2, fuel_3], axis=0)
all_fuel
```

```python
all_fuel = pd.concat([fuel_1, fuel_2, fuel_3], axis=1)
all_fuel
```
